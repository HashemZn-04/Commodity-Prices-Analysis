```python
# Importing packages
import pandas as pd
import numpy as np
import seaborn as sns
import datetime

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.dates as mdates
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

from statsmodels.tsa.seasonal import seasonal_decompose

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Dataset
df = pd.read_csv(r'C:\Users\hashe\OneDrive\Documents\Southampton Uni\Summer Projects\Python analysis project\commodity_prices.csv')
```

```python
df
```

```python
# Assessing for missing data via percentage of missing values in each column in the dataframe (df)

for col in df.columns: 
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}'.format(col, round(pct_missing*100)))
```

```python
# Viewing data types for all columns

print(df.dtypes)
```

```python
# Cleaning out duplicate entries

df.drop_duplicates()
```

```python
# Time series of dubai oil prices
plt.figure(figsize=(12, 6))

df['oil_dubai'].plot()

plt.title('Oil prices of dubai from 1960 to 2022')
plt.xlabel('Time')
plt.ylabel('Dubai oil prices')

plt.show()
```

```python
# Time series of UK and Norway brent oil prices

plt.figure(figsize=(12, 6))

df['oil_brent'].plot()

plt.title('Oil prices of UK and Norway from 1960 to 2022')
plt.xlabel('Time')
plt.ylabel('UK and Norway oil prices')

plt.show()
```

```python
# Correlation between brent oil and dubai oil prices

sns.regplot(x='oil_dubai', y='oil_brent', data=df)

plt.title('Relationship between dubai and brent oil prices of UK and Norway')
plt.xlabel('Dubai oil prices')
plt.ylabel('Brent oil prices')

plt.show()
```

```python
# A column representing average tea prices for all 3 tea types.

df['tea_avg'] = df[['tea_columbo', 'tea_kolkata', 'tea_mombasa']].mean(axis=1)

df.head()
```

```python
# Relationship between average tea prices and global sugar prices over time

plt.figure(figsize=(12, 6))

plt.plot(df.index, df['tea_avg'], label='Average Tea Price')
plt.plot(df.index, df['sugar_world'], label='Sugar Price')

plt.title('Time series of Average Tea Prices and Sugar Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

plt.show()
```

```python
# Relationship between average tea and average coffee prices

df['tea_avg'] = df[['tea_columbo', 'tea_kolkata', 'tea_mombasa']].mean(axis=1)
df['coffee_avg'] = df[['coffee_arabica', 'coffee_robustas']].mean(axis=1)

sns.regplot(x='tea_avg', y='coffee_avg', data=df)

plt.title('Relationship between average tea and average coffee prices')
plt.xlabel('Average tea prices')
plt.ylabel('Average coffee prices')

plt.show()
```

```python
# Pearson correlation method between all numeric variables

numeric_df = df.select_dtypes(include=[np.number])

correlation_matrix = numeric_df.corr(method ='pearson')

print(correlation_matrix)
```

```python
# Kendall correlation method between all numeric variables

numeric_df = df.select_dtypes(include=[np.number])

correlation_matrix = numeric_df.corr(method ='kendall')

print(correlation_matrix)
```

```python
# Spearman correlation method between all numeric variables

numeric_df = df.select_dtypes(include=[np.number])

correlation_matrix = numeric_df.corr(method ='spearman')

print(correlation_matrix)
```

```python
# Heatmap of a spearman correlation matrix

correlation_matrix = numeric_df.corr(method ='spearman')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation matrix of commodity prices')
plt.xlabel('Commodity correlation')
plt.ylabel('Commodity correlation')

plt.show()
```

```python
# growth rate of oil prices 
growth_rate1 = df['oil_dubai'].pct_change()
growth_rate1 = growth_rate1 * 100

dubai_oil_growth = f"{'oil_dubai'}_growth_rate"

df[dubai_oil_growth] = growth_rate1

growth_rate2 = df['oil_brent'].pct_change()
growth_rate2 = growth_rate2 * 100

brent_oil_growth = f"{'oil_brent'}_growth_rate"

df[brent_oil_growth] = growth_rate2

df
```
